```python
# econometrics5 pg.95
# 자기회귀(AR)모형: 과거 자기 변수가 미래의 자기 변수에 영향을 미치는 관계

# 만약 현재의 시계열이 현재 및 과거 오차항에 의해서만 영향을 받는다고 하면,
# 오차항이 미래의 변수에 영향을 미치는 관계: 이동평균(MA) 모형

# 자기회귀 모형과 이동평균 모형을 결합하면: 자기회귀이동평균(ARMA) 모형
# AR, MA, ARMA에서 가정하는 오차항은 백색잡음(white noise) 오차항임.
# 따라서 오차항에 대한 검증을 해봤을 때 white noise 오차항이 아니라면 다시 분석해야 함

# econometrics5 pg. 96
#시계열 안정성 판단기준(척도): 자기상관함수(ACF), 편자기상관함수(PACF)
#에르고딕성 (pg.96, 97)
# AR/MA 안정성 판단할 때 Ergodic성을 판단함
# -> 시계열의 안정성 테스트할 때 모든 자기공분산의 합이 특정값으로 수렴하는가 아니면 발산하는가를 판단

# 이동평균(MA) 모형
# 만약 과거 오차항이 현재 시계열에 영향을 미치는 구조이며, 함수형태가 선형함수라면: q차 이동평균모형 (MA(q))

# white noise 개념
# - 만약 y_t = e_t 에서 e_t가 white noise라면 y_t는 전혀 예측 불가한 random 변수의 형태를 띄고 있다는 뜻 
# (추세x, 분산일정, cycle x, 즉 완전히 random)

#econometrics5 pg. 98, 99
# MA(1): Y_t = u + e_t + beta_1*e_t-1
# MA(q): Y_t = u + e+t + beta_1*e_t-1 + beta2*e_t-2 + ... + beta_q*e_t-q
# MA(1) 모형에서 오차항이 백색잡음(white noise)일 경우 1차 이동평균모형은 안정적 확률과정이며 1차적률에 대한 에르고딕 과정임
# MA(q) = 약안정성, 에르고딕성 만족 -> 시계열분석에서 사용 가능한 모형 -> 확률과정(Y_t)의 예측값으로 사용 가능
# (pg.100) 정리하자면, MA(1), MA(q)는 특별한 조건 없더라도 약안정성/에르고딕성 만족

# econometrics5 pg. 99
# 자기회귀: 자기변수의 과거 값들이 미래의 자기변수 값의 평균 수준으로 회귀하는 현상 AR(p)
# pg.101  MA의 경우 오차항의 영향만 고려 (임의의 알수없는 충격만 점점 고차시계열로 영향 미침)
# AR의 경우 과거의 오차항 뿐 아니라 그 전기의 y값과 상수항(a0)의 영향을 모두 Y_t에 영향 미치는 변수로 봄.
# 따라서 AR이 MA보다는 좀 더 많은 시계열의 영향력을 고려한 모형이다 라고 할 수 있음

# econometrics5 pg. 104
# AR(1)의 경우:
# if |a1| < 1: stationary
# if |a1| > 1: not stationary
# if |a1| = 1: random walk process (대표적인 불안정 시계열)

#pg.105
#즉, if |a1| < 1: 평균과 분산에 대하여 약안정성 조건 만족
#pg. 107
# [정리] AR(1)에서 if |a1| < 1: 약안정성 조건 만족, 즉 |a1| < 1인 경우 AR(1)은 weak stationary 모형이다
# if |a1| >= 1: 불안정 시계열 모형
# AR(1): Y_t = a0 + a1*Y_t-1 + E_t

#AR(p)의 경우 |lambda_vector| < 1 만족시 (inside unit circle) 약안정성 조건 만족 (pg.132)

# econometrics5 pg.133, 134
# AR모형 - 충격반응분석 (Impulse Response Analysis)
# 충격반응분석: j기 전 오차항에 의한 충격이 발생했을 때 각 시점별 영향력을 측정하는 분석방법
# 목적: 1) 충격의 지속시간 분석, 2) 충격의 정도 분석
# 오차충격(E_t-j)이 발생햇을 때 Y_t-j...Y_t...Y_t+2... 까지 각 시점별로 얼마나 영향을 미치는지 측정
# 외부에서 발생한 외생적 충격이 시계열 변수에 미치는 영향(영향의 크기 및 지속시간) 등 분석에 활용

# econometrics5 pg. 135
# AR은 자기 변수 과거값 -> 현재에 영향을 미치는 구조
# 이 때 오차항은 자기변수 과거값에 의해 설명되지 않는 나머지 부분
# 즉 이 시계열이 아니라 외부적인 영향에 의한 충격이라 할 수 있음 (외부의 외생적 충격)
# 이 충격이 시계열에 미치는 영향 분석(크기, 지속시간)이 IRA
# pg.148 -> AR(p)의 경우 IRA

# econometrics5 pg.149
# 시계열의 안정성 검정 - 자기상관 검정
# ACF: 직/간접 영향 모두 포함
# PACF: 직접 영향만 (간접 영향은 통제/제거)
# 자기상관검정: 자기상관함수와 편자기상관함수를 통해 시계열의 안정성을 검정하는 방법
# 안정적 시계열이라면, 자기상관함수와 편자기상관함수가 장기적으로 0에 수렴 (시계열 차이가 길어질수록 시계열 간 상관이 낮아짐)
# 자기상관검정방법: t-1기에서 시작하여 순차적으로 표본 ACF와 표본 PACF가 0인지를 검정
# - 검정 전에 그래프를 통해 표본 ACF와 표본 PACF가 장기적으로 감소하며 0에 수렴하는지 확인해야 함
```
