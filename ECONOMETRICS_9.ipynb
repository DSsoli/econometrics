```python
# 3.2. White noise
# White noise has...

# Constant mean
# Constant variance
# Zero auto-correlation at all lags
```

```python
import numpy as np
# Plotting white noise
rcParams['figure.figsize'] = 16, 6
white_noise = np.random.normal(loc=0, scale=1, size=1000)
# loc is mean, scale is variance
plt.plot(white_noise)
```

```python
# Plotting autocorrelation of white noise
plot_acf(white_noise,lags=20)
plt.show()
# See how all lags are statistically insigficant as they lie inside the confidence interval(shaded portion).
```

```python
# 3.3. Random WalkÂ¶
# A random walk is a mathematical object, known as a stochastic or random process, 
# that describes a path that consists of a succession of random steps on some mathematical space such as the integers.

# In general if we talk about stocks, Today's Price = Yesterday's Price + Noise

# Pt = Pt-1 + Îµt
# Random walks can't be forecasted because well, noise is random.

# Random Walk with Drift(drift(Î¼) is zero-mean)

# Pt - Pt-1 = Î¼ + Îµt

# Regression test for random walk

# Pt = Î± + Î²Pt-1 + Îµt
# Equivalent to Pt - Pt-1 = Î± + Î²Pt-1 + Îµt

# Test:

# H0: Î² = 1 (This is a random walk)
# H1: Î² < 1 (This is not a random walk)

# Dickey-Fuller Test:

# H0: Î² = 0 (This is a random walk)
# H1: Î² < 0 (This is not a random walk)
# Augmented Dickey-Fuller test
# An augmented Dickeyâ€“Fuller test (ADF) tests the null hypothesis that a unit root is present in a time series sample. 
# It is basically Dickey-Fuller test with more lagged changes on RHS.
```

```python
# Augmented Dickey-Fuller test on volume of google and microsoft stocks 

adf = sm.tsa.adfuller(microsoft["Volume"])
print("p-value of microsoft: {}".format(float(adf[1])))
adf = sm.tsa.adfuller(google["Volume"])
print("p-value of google: {}".format(float(adf[1])))

# As microsoft has p-value 0.0003201525 which is less than 0.05, null hypothesis is rejected and this is not a random walk.
# Now google has p-value 0.0000006510 which is more than 0.05, null hypothesis is rejected and this is not a random walk.
```

```python
#Generating a random walk
from numpy.random import normal, seed
seed(42)
rcParams['figure.figsize'] = 16, 6
random_walk = normal(loc=0, scale=0.01, size=1000)
plt.plot(random_walk)
plt.show()
```

```python
# Stationarity
# Stationarity is important as non-stationary series that depend on time have too many parameters to account for 
# when modelling the time series. diff() method can easily convert a non-stationary series to a stationary series.

# We will try to decompose seasonal component of the above decomposed time series.
```

```python
# The original non-stationary plot
decomposed_google_volume = sm.tsa.seasonal_decompose(google["High"],period=365)
decomposed_google_volume.trend.plot()
```

```python
# The new stationary plot
decomposed_google_volume.trend.diff(1).plot()
```

```python
ar1 = np.array([1, -0.9]) # We choose -0.9 as AR parameter is +0.9
ma1 = np.array([1])
AR1 = sm.tsa.ArmaProcess(ar1, ma1)
sim1 = AR1.generate_sample(nsample=1000)
```

```python
#Simulating AR(1) model
model = sm.tsa.ARIMA(sim1, order=(1,0,0))
result = model.fit()
print(result.summary())
print("Î¼={} ,Ï•={}".format(result.params[0],result.params[1]))
```

```python
# Ï• is around 0.9 which is what we chose as AR parameter in our first simulated model.
```

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_predict
#plt.plot(sim1[900:], label='y')
#plt.plot(result.predict(start=900, end=1010), label='predict')
fig, ax = plt.subplots()
ax = pd.Series(sim1[900:]).plot(ax=ax)
plot_predict(result, start=900, end=1010, ax=ax)
plt.show()
```

```python
plt.plot(sim1[900:], label='y')
plt.plot(result.predict(start=900, end=1010), label='prediction')
plt.legend()
plt.show()
```
